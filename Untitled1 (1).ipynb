```python
# sunspots_attention_forecasting.py

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.metrics import mean_squared_error, mean_absolute_error
import tensorflow as tf
from tensorflow.keras.layers import Input, LSTM, GRU, Dense, Layer, concatenate
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping
from statsmodels.tsa.arima.model import ARIMA

# 1. Load and Preprocess Sunspots Data

from statsmodels.datasets import sunspots

def load_sunspots():
    dataset = sunspots.load()
    # statsmodels sunspots data typically starts in 1700-01
    # Use dataset.endog which is the time series values
    series_values = dataset.endog
    # Create a DatetimeIndex for the sunspot data, starting from Jan 1700
    start_date = '1700-01-01'
    dates = pd.date_range(start=start_date, periods=len(series_values), freq='MS') # 'MS' for Month Start
    # Create the final DataFrame with 'Month' and 'Sunspots' columns
    df = pd.DataFrame({'Month': dates, 'Sunspots': series_values})
    return df

df = load_sunspots()
values = df['Sunspots'].values.reshape(-1, 1)

sch_scaler = MinMaxScaler()
scaled = sch_scaler.fit_transform(values)

def create_sequences(data, look_back=24):
    X, y = [], []
    for i in range(len(data) - look_back):
        X.append(data[i:i+look_back])
        y.append(data[i+look_back])
    return np.array(X), np.array(y)

LOOK_BACK = 24
X, y = create_sequences(scaled, look_back=LOOK_BACK)
split = int(0.8 * len(X))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 2. Custom Attention Layer

class BahdanauAttention(Layer):
    def __init__(self, units):
        super(BahdanauAttention, self).__init__()
        self.W1 = Dense(units)
        self.W2 = Dense(units)
        self.V = Dense(1)

    def call(self, query, values):
        score = self.V(tf.nn.tanh(self.W1(query) + self.W2(values)))
        attention_weights = tf.nn.softmax(score, axis=1)
        context_vector = attention_weights * values
        context_vector = tf.reduce_sum(context_vector, axis=1)
        return context_vector, attention_weights


class SqueezeLayer(Layer):
    def __init__(self, axis=None, **kwargs):
        super(SqueezeLayer, self).__init__(**kwargs)
        self.axis = axis

    def call(self, inputs):
        return tf.squeeze(inputs, axis=self.axis)

    def get_config(self):
        config = super(SqueezeLayer, self).get_config()
        config.update({'axis': self.axis})
        return config


#  3. Build Attention-based Model (LSTM + Attention)

def build_attention_model(hidden_units=64, look_back=LOOK_BACK, lr=0.001):
    inputs = Input(shape=(look_back, 1))
    lstm_out = LSTM(hidden_units, return_sequences=True)(inputs)
    attention, weights = BahdanauAttention(hidden_units)(lstm_out, lstm_out)

    last_input_step = inputs[:, -1, :]
    concat = concatenate([attention, last_input_step], axis=-1)

    dense = Dense(1)(concat)
    model = Model(inputs=inputs, outputs=dense)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
                  loss='mse',
                  metrics=['mae'])
    return model

#  4. Baseline LSTM Model

def build_baseline_model(hidden_units=64, look_back=LOOK_BACK, lr=0.001):
    inputs = Input(shape=(look_back, 1))
    lstm_out = LSTM(hidden_units)(inputs)
    dense = Dense(1)(lstm_out)
    model = Model(inputs=inputs, outputs=dense)
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=lr),
                  loss='mse',
                  metrics=['mae'])
    return model

def grid_search(X_train, y_train, X_val, y_val):
    best_mae = float('inf')
    best_config = {}
    for units in [32, 64, 128]:
        for lr in [0.01, 0.001]:
            model = build_attention_model(hidden_units=units, look_back=LOOK_BACK, lr=lr)
            es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
            model.fit(X_train, y_train, validation_data=(X_val, y_val), epochs=50, batch_size=32, callbacks=[es], verbose=0)
            val_mae = model.evaluate(X_val, y_val, verbose=0)[1]
            if val_mae < best_mae:
                best_mae = val_mae
                best_config = {'units': units, 'lr': lr}
    return best_config
val_split = int(0.9 * len(X_train))
best_config = grid_search(X_train[:val_split], y_train[:val_split], X_train[val_split:], y_train[val_split:])
print("Best Attention Model Config:", best_config)


attention_model = build_attention_model(hidden_units=best_config['units'], look_back=LOOK_BACK, lr=best_config['lr'])
baseline_model = build_baseline_model(hidden_units=best_config['units'], look_back=LOOK_BACK, lr=best_config['lr'])

es = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

attention_model.fit(X_train, y_train, validation_split=0.1, epochs=100, batch_size=32, callbacks=[es], verbose=0)
baseline_model.fit(X_train, y_train, validation_split=0.1, epochs=100, batch_size=32, callbacks=[es], verbose=0)

# 5. Evaluation

def evaluate_model(model, X_test, y_test, scaler):
    preds = model.predict(X_test)
    preds_rescaled = scaler.inverse_transform(preds)
    y_test_rescaled = scaler.inverse_transform(y_test)
    rmse = np.sqrt(mean_squared_error(y_test_rescaled, preds_rescaled))
    mae = mean_absolute_error(y_test_rescaled, preds_rescaled)
    mape = np.mean(np.abs((y_test_rescaled - preds_rescaled) / y_test_rescaled)) * 100
    return rmse, mae, mape, preds, y_test_rescaled

att_rmse, att_mae, att_mape, att_preds, y_test_rescaled = evaluate_model(attention_model, X_test, y_test, sch_scaler)
base_rmse, base_mae, base_mape, base_preds, y_test_rescaled = evaluate_model(baseline_model, X_test, y_test, sch_scaler)

#  6. Textual Interpretation and Attention Visualization

def extract_attention_weights(model, X):

    pass  # For full implementation, see framework-specific docs

# Plot predictions vs actual
plt.figure(figsize=(12, 6))
plt.plot(y_test_rescaled, label="True")
plt.plot(att_preds, label="Attention Model")
plt.plot(base_preds, label="Baseline LSTM")
plt.legend()
plt.title("Sunspots Forecasting: Actual vs Predictions")
plt.show()
```
